```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
%matplotlib inline
```

```python
df=pd.read_excel("D:\\Superstore.xls")
df
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#boxplot
data=[df['Sales'],df['Profit'],df['Quantity'],df['Discount']]
fig=plt.figure(figsize=(9,6))
ax=fig.add_subplot(111)
plot=ax.boxplot(data)
```

```python
# 1st segment and ship mode wise profit
seg_ship=df.groupby(["Segment","Ship_Mode"]).mean()[["Profit"]]
seg_ship
```

```python
fig,ax=plt.subplots(figsize=(9,6))
seg_ship.unstack().plot(ax=ax)
```

```python
# 2nd month wise profit and sales
df['month']=pd.DatetimeIndex(df['Order_Date']).month
df['month']
```

```python
prof_sales=df.groupby(['month']).mean()[['Profit','Sales']]
prof_sales
```

```python
style.use("ggplot")
ax=prof_sales.plot(kind='bar',figsize=(11,7),rot=60)
ax.set_title("Monthwise Profit & Sales")
plt.show()
```

```python
# 5th scatter plot for quantity vs profit
style.use('seaborn-white')
plt.scatter(df["Quantity"],df["Profit"],alpha=0.5)
plt.xlabel('Profit')
plt.ylabel('Quantity')
plt.show()
```

```python
# 3rd top 10 cities based on profit
df_sorted=df.sort_values("Profit",ascending=False)
df_sorted.head(10)
```

```python
df_sorted.head(10).describe()
```

```python
x=df_sorted.head(10)['City']
y=df_sorted.head(10)['Profit']
print(x,y)
```

```python
style.use('ggplot')
plt.figure(figsize=(11,8))
plt.bar(x,y)
plt.xlabel('City')
plt.ylabel('Profit')
plt.title('profit of top 10 cities')
plt.show()
```

```python
# 4th region-wise top 5 cities with graph
reg_city=df.groupby(["Region","City"]).mean()[["Profit"]]
reg_city.reset_index(inplace=True)
east=reg_city[reg_city['Region']=='East'].sort_values(by=['Profit'],ascending=False).head(5)
plt.bar(east['City'],east['Profit'],color='r')
plt.title("Top 5 Cities of East region") 
plt.xlabel("East") 
plt.ylabel("Profit") 
```

```python
south=reg_city[reg_city['Region']=='South'].sort_values(by=['Profit'],ascending=False).head(5)
plt.bar(south['City'],south['Profit'],color='b')
plt.title("Top 5 Cities of South region") 
plt.xlabel("South") 
plt.ylabel("Profit") 
```

```python
West=reg_city[reg_city['Region']=='West'].sort_values(by=['Profit'],ascending=False).head(5)
plt.bar(West['City'],West['Profit'],color='g')
plt.title("Top 5 Cities of West") 
plt.xlabel("West") 
plt.ylabel("Profit") 
```

```python
Central=reg_city[reg_city['Region']=='Central'].sort_values(by=['Profit'],ascending=False).head(5)
plt.bar(Central['City'],Central['Profit'],color='y')
plt.title("Top 5 Cities of West") 
plt.xlabel("Central") 
plt.ylabel("Profit") 
```
